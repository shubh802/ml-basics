```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) # creates the figure
# subplot specifies numrows, numcols, fignum (range 1 to numrows * numcols)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# concatinate
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "temperature": [32, 45, 30],
    "humidity": [45, 75, 81]
})

india_weather

us_weather = pd.DataFrame({
    "city":["new york", "chicago", "orlando"],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})

us_weather
```

```python
df = pd.concat([india_weather, us_weather], ignore_index = True)
df
```

```python
df = pd.concat([india_weather, us_weather], axis=1)
df
```

```python
# Merge
temp_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore", "hyderabad"],
    "temperature": [32, 45, 30, 40]
})

temp_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "humidity":[47, 84, 56]
})
humidity_df
```

```python
df = pd.merge(temp_df, humidity_df, on="city")
df
```

```python
df = pd.merge(temp_df, humidity_df, on="city", how="outer")
df
```
